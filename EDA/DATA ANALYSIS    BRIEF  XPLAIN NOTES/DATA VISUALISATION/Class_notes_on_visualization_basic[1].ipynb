### 1. Line chart

- Line plot is a type of chart that displays information as a series of data points connected by straight line segments

- Note: Line Charts are best suited in displaying trends and patterns of variables in data.

`Note: Data Should be Numerical Data`

```python
x = np.arange(10)
print("The Values of x is", x)
y = x ** 2
print("The Values of y is", y)

#lets plot x,y 
plt.plot(x, y) # Plot y versus x as lines and/or markers.

plt.show() #Display all open figures.
```

```python
plt.plot(x, y, 'b+') # plot x and y using blue + markers

plt.show() #Display all open figures.
```

### Refeence: on plt.plot() -->  https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
plt.plot([1, 3, 5, 7, 9], [1, 8, 3, 5, 1],"b+")

plt.xlabel('x')  # label x axis in the plot
plt.ylabel('y') # label y axis in the plot
plt.title('Line Plot - x & y') #add title to the plot

plt.xlim([0, 10])  #Get or set the x limits of the current axes.
plt.ylim([0, 10])  #Get or set the y limits of the current axes.

plt.show()
```

```python
# mobile sales by year

# generate 4 arrays with 12 random values each 

np.random.seed(111)
sales2017 = np.round(np.sort(np.random.uniform(15,55,size=12 )), 2)
sales2018 = np.round(np.sort(np.random.uniform(25,75,size=12 )), 2)
sales2019 = np.round(np.sort(np.random.uniform(9,25,size=12)), 2)
sales2020 = np.round(np.sort(np.random.uniform(20,35,size=12)), 2)
```

```python
month_num = np.arange(1,13, 1)
print(month_num)

plt.figure(figsize=(12,5), dpi = 100)

plt.plot(month_num, sales2017, label = '2017 Sales', color = 'b', ls = '-', lw = 2, marker='.', markersize = 12)
plt.plot(month_num, sales2018, label = '2018 Sales', color = 'orange', ls = '--', lw = 3, marker='o', markersize = 12)
plt.plot(month_num, sales2019, label = '2019 Sales', color = 'lime', ls= '-.', lw = 3, marker='D', markersize = 12)
plt.plot(month_num, sales2020, label = '2020 Sales', color = '#fd3c06', ls = ':', lw = 4, marker='H', markersize = 12)

plt.xlabel('Month Number')
plt.ylabel('Sales ')

plt.title('Monthly Mobile Store Sales by Year')

plt.xticks(np.arange(1,14, 1))
plt.yticks(np.arange(5,100, 15))

plt.legend();
```

### 2. Scatter Plots

- Scatter plot is a graph in which the values of two variables are plotted along two axes. It is a most basic type of plot that helps you visualize the relationship between two variables.

- The Scatter Plot Charts are ideal in situations where you’ve got the random distribution of variables in data.

- If your goal is to display insights into hidden correlations and associations present in your data, your go-to visualization design should be a Scatter Plot Chart.

`Note: Both X and Y Variables are the Numerical Data` 

```python
np.random.seed(123)

temp = np.random.uniform(21,50, size =14).round(2)
AC_sales = np.sort(np.random.randint(1000, 10000, size=14))
TV_sales = np.random.randint(5000, 10000, size=14)
```

```python
temp 
```

```python
AC_sales 
```

```python
TV_sales 
```

```python
plt.plot(temp, AC_sales, marker = 'D',)

plt.xlabel('Temp', fontsize=16)
plt.ylabel('AC Sales', fontsize=16)
plt.title('Scatter plot - Temp vs AC Sales',fontsize=20)
plt.show();
```

```python
plt.scatter(temp, AC_sales, marker = 'D', s = 50, c = 'green')

plt.xlabel('Temp', fontsize=12)
plt.ylabel('AC Sales', fontsize=16)
plt.title('Scatter plot - Temp vs AC Sales',fontsize=20)
plt.show()
```

#### Scatter plot by categories

```python
import matplotlib.pyplot as plt
np.random.seed(123)

temp = np.random.uniform(21,50, size =14).round(2)
AC_sales = np.sort(np.random.randint(1000, 10000, size=14))
TV_sales = np.random.randint(5000, 10000, size=14)
```

```python
plt.scatter(temp, AC_sales, marker = 'D', s = 50, c = 'teal', label = 'AC')
plt.scatter(temp, TV_sales, marker = 'H', s = 50, c = 'red', label = 'TV')


plt.xlabel('Temp', fontsize=16)
plt.ylabel('Sales', fontsize=16)
plt.title('Scatter plot - Temp vs AC & TV Sales',fontsize=15)
plt.legend()
plt.show();
```

### 3. Boxplot

Make a box and whisker plot for each column of x or each vector in sequence x. The box extends from the lower to upper quartile values of the data, with a line at the median. The whiskers extend from the box to show the range of the data. Flier points are those past the end of the whiskers.

 - A boxplot summarizes the distribution of a numeric variable for one or several groups. It allows to quickly get the median, quartiles and outliers but also hides the dataset individual data points.

```python
# Data: Salaries in lacs
np.random.seed(123)

fresher = np.round(np.random.uniform(3.5, 6.5, size = 10),1)
lead = np.round(np.random.uniform(5.5, 12.5, size = 10),1)
manager = np.round(np.random.uniform(8.5, 16.5, size = 10),1)
director = np.round(np.random.uniform(18.25, 34, size = 10),1)
vp = np.round(np.random.uniform(27, 50.5, size = 10),1)
cxo = np.round(np.random.uniform(45.5,75, size = 10),1)
```

```python
fresher
```

```python
lead 
```

```python
manager
```

```python
fresher
```

```python
plt.figure(figsize=(7,3))
plt.boxplot(fresher, vert = 1); # vert = 0 for horizontal box plot
```

## 4. Pie Chart



```python
Mfg = ['Mahindra','Toyota','Renault','Honda']
Vehicles_sold = [155.54,93.12,92.27,82.07]

plt.figure(figsize=(10,10), dpi=100)

plt.pie(Vehicles_sold, labels=Mfg, autopct='%.1f%%')
#plt.pie(Vehicles_sold, labels=Mfg)

plt.title('Car Market share - India 2021')
plt.legend();
```
