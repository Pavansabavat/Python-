### 3.1 Categorical Scatterplots - stripplot()

```python
sns.stripplot(data = tips, x = 'day', y='total_bill')
```

```python
sns.stripplot(data = tips, x = 'day', y='tip')
```

### 3.2 Swarm Plot

This function is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they dont overlap. This gives a better representation of the distribution of values, but it does not scale well to large numbers of observations. This style of plot is sometimes called a beeswarm.

```python
sns.swarmplot(data = tips, x='day', y='total_bill')
```

### 3.3 Boxplot

```python
sns.boxplot(data = tips, x='day', y='total_bill')
```

```python
tips.loc[tips.time == 'Lunch', :]
```

### Catplot


If you are working with data that involves any categorical variables.a data visualization library Seaborn encompasses several types of categorical plots into a single function: catplot()

catplot() function includes 8 different types of categorical plots. The three most common plots: count plots, bar plots, and box plots.


Categorical scatterplots with catplot  

stripplot()  with kind=strip  
swarmplot()  with kind=swarm  

Categorical distribution plots with catplot  

boxplot()  with kind=box  
violinplot()  with kind=violin  
boxenplot()  with kind=boxen  
Categorical estimate plots with catplot  
pointplot()  with kind=point  
barplot()  with kind=bar  
countplot()  with kind=count  

```python
sns.catplot(x="Gender", y="ApplicantIncome", data=df_loan)
```

```python
sns.catplot(x="Gender", y="ApplicantIncome", data=df_loan,kind='box')
```

```python
sns.catplot(x='Gender', data=df_loan, kind='count')
```

Changing the order of categories

In our plot, the quality of the cut is from best to worst. But lets reverse the order: 

```python
sns.catplot(x='Property_Area', data=df_loan, kind='count')
```

```python
category_order = ['Rural', 'Urban', 'Semiurban']

sns.catplot(x='Property_Area', data=df_loan, kind='count', order=category_order);
```

It is also important to keep in mind that a bar plot shows only the mean (or other estimator) value as default

```python
sns.catplot(x='Item_Type',
            y='Total_Profit',
            data=df_sales,
            kind='bar',
            ci=None)

plt.xticks(rotation = 90)
plt.show()
```

```python
sns.catplot(x='Item_Type', y='Total_Profit', data=df_sales, kind='box')  
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.catplot(x='Sales_Channel', y='Total_Profit',data=df_sales,kind="box")
```

```python
sns.catplot(x='Sales_Channel', y='Total_Profit',data=df_sales,kind="violin")
```

### What is the difference between a bar chart and a histogram?   

The major difference is that a histogram is only used to plot the frequency of score occurrences in a continuous data set that has been divided into classes, called bins. Bar graph is a graphical representation of data that uses bars from numerical variable to compare different categories of categorical variable.

Histogram used for displaying the frequency of occurrences and Bar Plots helps you to compare different categories of data.
