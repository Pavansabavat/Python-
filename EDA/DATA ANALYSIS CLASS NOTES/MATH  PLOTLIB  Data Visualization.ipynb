## Line chart

```python
x=np.arange(10)
print("the vale of x is",x)
y=x**2
print("the vale of y is ",y)

plt.plot(x,y) ## to make graph using x  &  y

plt.show()  ## plt.show  to show all the figures in graph
```

```python
x=np.arange(-9, 10)
print("the vale of x is",x)
y=x**2
print("the vale of y is ",y)

plt.plot(x,y) ## to make graph using x  &  y

plt.show()
```

```python
plt.plot(x,y,'b+') ## plot x and y using  blue+ markers
plt.show
```

```python
plt.plot(x,y,'r>') ## plot x and y using r- markers
plt.show
```

```python
plt.plot([1,3,8,6,7],[2,5,6,7,8])

plt.xlabel('x=axis') ## label x axis in the plot
plt.ylabel('y-axis') ## label y axis in the plot
plt.title('Linee Plot - x & y') ## add title to the plot

plt.xlim([0,10]) ## get or set the x limits of current axis
plt.ylim([0,10]) ## get or set the y limits of current axis

plt.show()
```

```python
np.random.seed(111)   ## seed is used to get the saved output   othrwise  if u not mention it will gives random outputs
sales2017 = np.round(np.sort(np.random.uniform(15,55,size=12)),2)
sales2018 = np.round(np.sort(np.random.uniform(25,64,size=12)),2)
sales2019 = np.round(np.sort(np.random.uniform(22,34,size=12)),2)
sales2020 = np.round(np.sort(np.random.uniform(20,35,size=12)),2)
```

```python
len(sales2018)
```

```python
sales2019
```

```python
month=np.arange(1,13,1)
print(month)

plt.figure(figsize=(18,7),dpi=100)

plt.plot(month,sales2017,label='2017 sales',color='r', ls='-',lw=2,marker='.',markersize=12) ##  ls= length of size, lw=width 
plt.plot(month,sales2018,label='2018 sales',color='orange', ls='--',lw=2,marker='o',markersize=12)
plt.plot(month,sales2019,label='2019 sales',color='lime', ls='-.',lw=2,marker='>',markersize=12)
plt.plot(month,sales2020,label='2020 sales',color='yellow', ls=':',lw=4,marker='h',markersize=12)

plt.xlabel("month Number")
plt.ylabel('sales')

plt.title("Monthly mobile store sales by year")

plt.xticks(np.arange(1,11,1))
plt.yticks(np.arange(5,70,10))

plt.legend(loc='best');
```

## pie chart
 to compare shares 

```python
brand=["mahindra","toyota","renault","honda"]
vehicles_sold=[87.8,54.9,45.35,79.35]

plt.figure(figsize=(10,10),dpi=100)
plt.pie(vehicles_sold,labels=brand,autopct='%.2f%%')## autopct='%.2f%%' 2 means 2 decimals(it means percentage  displa on graph)

plt.title('car market share-india 2021')
plt.legend();
```

## 5 Bar chart
 to compare indivigal  data

```python
brand=["mahindra","toyota","renault","honda"]
vehicles_sold=[87.8,66.2,45.35,79.35]
plt.bar(brand,vehicles_sold,color='green',edgecolor='black')
plt.show
```

### Pie charts represent data in a circle, with “slices” corresponding to percentages of the whole,
### whereas bar graphs use bars of different lengths to represent data in a more flexible way

```python
brand=["mahindra","toyota","renault","honda"]
vehicles_sold=[87.8,66.2,45.35,79.35]
plt.bar(brand,vehicles_sold,color='green',edgecolor='black')
plt.show
```
