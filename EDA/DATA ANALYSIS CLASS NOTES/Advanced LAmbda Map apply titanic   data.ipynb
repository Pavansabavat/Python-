```python
df["Name1"]=df["Name"].str.split(",")
df.head()
```

```python
df[["Last","First"]]=df["Name"].str.split(",",expand=True)
df.head()
```

```python
for i in range (len(df)):
    s=df.loc[i,"Sex"]
    if s=="male":
        df.loc[i,"Gender"]=0
    else:
        df.loc[i,"Gender"]=1
df.head()        
```

```python
df["gender_map"]=df["Sex"].map(new)
df.head()
```

### using  function

```python
def newGender(x):
    if x=="male":
        return 0
    else:
        return 1
```

```python
newGender("female")
```

```python
for i in range (len(df)):
    s=df.loc[i,"Sex"]
    df.loc[i,"Gender_function"]=newGender(s)
df.head()
```

```python
df["Gender_apply"]=df["Sex"].apply(newGender)
df.head()
```

```python
df["Gender_Lambda"]=df["Sex"].apply(lambda x:0 if x=="male" else 1)
df.head()
```
